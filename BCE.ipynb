```python
# --- 1. SETUP AND DATA DOWNLOAD ---
!mkdir -p ~/.kaggle
!cp kaggle.json ~/.kaggle/
!kaggle datasets download -d fakhrealam9537/leaf-disease-segmentation-dataset

import zipfile
import os
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, models
from tqdm import tqdm

# Extract Dataset
zip_ref = zipfile.ZipFile('/content/leaf-disease-segmentation-dataset.zip', 'r')
zip_ref.extractall('/content')
zip_ref.close()

# --- 2. CONFIGURATION ---
DATASET_PATH = "/content/aug_data/aug_data"
IMG_SIZE = (256, 256)
BATCH_SIZE = 8
LEARNING_RATE = 5e-5
NUM_EPOCHS = 140
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# --- 3. LOSS IMPLEMENTATION ---
criterion = nn.BCEWithLogitsLoss()

# --- 4. DATASET CLASS ---
class LeafDiseaseDataset(Dataset):
    def __init__(self, root_dir, img_size=IMG_SIZE, augment=False):
        self.images_dir = os.path.join(root_dir, "images")
        self.masks_dir = os.path.join(root_dir, "masks")
        self.img_size = img_size
        self.image_files = sorted([f for f in os.listdir(self.images_dir) if f.endswith('.jpg')])
        self.augment = augment
        self.normalize = transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                              std=[0.229, 0.224, 0.225])

    def __len__(self): return len(self.image_files)

    def __getitem__(self, idx):
        img_name = self.image_files[idx]
        img_path = os.path.join(self.images_dir, img_name)
        mask_path = os.path.join(self.masks_dir, img_name.replace(".jpg", ".png"))

        image = Image.open(img_path).convert("RGB")
        mask = Image.open(mask_path).convert("L")

        image = image.resize(self.img_size)
        mask = mask.resize(self.img_size, resample=Image.NEAREST)

        if self.augment:
            if np.random.rand() > 0.5:
                image = transforms.functional.hflip(image)
                mask = transforms.functional.hflip(mask)
            if np.random.rand() > 0.5:
                image = transforms.functional.vflip(image)
                mask = transforms.functional.vflip(mask)

        image = transforms.ToTensor()(image)
        image = self.normalize(image)

        mask = transforms.ToTensor()(mask)
        mask = (mask > 0).float()

        return image, mask

# --- 5. DATASET SPLIT ---
torch.manual_seed(42)
full_dataset = LeafDiseaseDataset(DATASET_PATH)
train_size = int(0.75 * len(full_dataset))

indices = torch.randperm(len(full_dataset))
train_indices = indices[:train_size]
val_indices = indices[train_size:]

train_dataset = torch.utils.data.Subset(
    LeafDiseaseDataset(DATASET_PATH, augment=True),
    train_indices
)

val_dataset = torch.utils.data.Subset(
    LeafDiseaseDataset(DATASET_PATH, augment=False),
    val_indices
)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=2, pin_memory=True)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)

# --- 6. MODEL ---
class DoubleConv(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 3, padding=1, bias=False),
            nn.BatchNorm2d(out_ch),
            nn.ReLU(inplace=True),
            nn.Dropout2d(0.2),
            nn.Conv2d(out_ch, out_ch, 3, padding=1, bias=False),
            nn.BatchNorm2d(out_ch),
            nn.ReLU(inplace=True)
        )
    def forward(self, x): return self.conv(x)

class DecoderBlock(nn.Module):
    def __init__(self, in_ch, skip_ch, out_ch):
        super().__init__()
        self.up = nn.ConvTranspose2d(in_ch, in_ch//2, 2, 2)
        self.conv = DoubleConv(in_ch//2 + skip_ch, out_ch)

    def forward(self, x, skip):
        x = self.up(x)
        x = torch.cat([x, skip], dim=1)
        return self.conv(x)

class ResNet50UNet(nn.Module):
    def __init__(self):
        super().__init__()
        base = models.resnet50(weights=models.ResNet50_Weights.DEFAULT)
        layers = list(base.children())

        self.init = nn.Sequential(*layers[:3])
        self.pool = layers[3]
        self.l1 = layers[4]
        self.l2 = layers[5]
        self.l3 = layers[6]
        self.l4 = layers[7]

        self.up1 = DecoderBlock(2048,1024,512)
        self.up2 = DecoderBlock(512,512,256)
        self.up3 = DecoderBlock(256,256,128)
        self.up4 = DecoderBlock(128,64,64)

        self.final_up = nn.ConvTranspose2d(64,32,2,2)
        self.final = nn.Conv2d(32,1,1)

    def forward(self,x):
        x0 = self.init(x)
        x1 = self.l1(self.pool(x0))
        x2 = self.l2(x1)
        x3 = self.l3(x2)
        x4 = self.l4(x3)

        d1 = self.up1(x4,x3)
        d2 = self.up2(d1,x2)
        d3 = self.up3(d2,x1)
        d4 = self.up4(d3,x0)

        return self.final(self.final_up(d4))

# --- 7. METRIC ---
def mean_iou(preds, masks, eps=1e-6):
    preds = (torch.sigmoid(preds) > 0.5).float()
    intersection = (preds * masks).sum((1,2,3))
    union = preds.sum((1,2,3)) + masks.sum((1,2,3)) - intersection
    return ((intersection+eps)/(union+eps)).mean()

# --- 8. TRAINING ---
model = ResNet50UNet().to(DEVICE)
optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)

history = {"train_loss": [], "val_loss": [], "train_iou": [], "val_iou": []}
best_iou = 0

for epoch in range(NUM_EPOCHS):
    model.train()
    t_loss, t_iou = 0, 0

    for images, masks in tqdm(train_loader, leave=False):
        images, masks = images.to(DEVICE), masks.to(DEVICE)

        outputs = model(images)
        loss = criterion(outputs, masks)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        t_loss += loss.item()
        t_iou += mean_iou(outputs, masks).item()

    model.eval()
    v_loss, v_iou = 0, 0

    with torch.no_grad():
        for images, masks in val_loader:
            images, masks = images.to(DEVICE), masks.to(DEVICE)

            outputs = model(images)
            v_loss += criterion(outputs, masks).item()
            v_iou += mean_iou(outputs, masks).item()

    history["train_loss"].append(t_loss/len(train_loader))
    history["val_loss"].append(v_loss/len(val_loader))
    history["train_iou"].append(t_iou/len(train_loader))
    history["val_iou"].append(v_iou/len(val_loader))

    print(f"Epoch {epoch+1}: "
      f"T-Loss: {history['train_loss'][-1]:.4f} | "
      f"V-Loss: {history['val_loss'][-1]:.4f} | "
      f"T-IoU: {history['train_iou'][-1]:.4f} | "
      f"V-IoU: {history['val_iou'][-1]:.4f}")

    if history["val_iou"][-1] > best_iou:
        best_iou = history["val_iou"][-1]
        torch.save(model.state_dict(), "best_model.pth")
        print("--->>>Best model saved!!")

# --- 9. PLOTS ---
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(history["train_loss"], label="Train Loss")
plt.plot(history["val_loss"], label="Val Loss")
plt.title("Loss Curve")
plt.legend()

plt.subplot(1,2,2)
plt.plot(history["train_iou"], label="Train IoU")
plt.plot(history["val_iou"], label="Val IoU")
plt.title("IoU Curve")
plt.legend()

plt.show()

# --- 10. VISUALIZATION ---
def visualize_results(model, dataset, n_samples=5):
    model.eval()
    fig, axes = plt.subplots(n_samples, 3, figsize=(15, 5*n_samples))

    for i in range(n_samples):
        idx = np.random.randint(len(dataset))
        image, mask = dataset[idx]

        with torch.no_grad():
            pred = model(image.unsqueeze(0).to(DEVICE))
            pred = (torch.sigmoid(pred) > 0.5).float().cpu().squeeze()

        img_disp = image.permute(1,2,0).numpy()
        img_disp = img_disp * np.array([0.229,0.224,0.225]) + np.array([0.485,0.456,0.406])
        img_disp = np.clip(img_disp,0,1)

        axes[i,0].imshow(img_disp)
        axes[i,0].set_title("Image")

        axes[i,1].imshow(mask.squeeze(), cmap='gray')
        axes[i,1].set_title("GT")

        axes[i,2].imshow(pred, cmap='gray')
        axes[i,2].set_title("Prediction")

        for ax in axes[i]:
            ax.axis("off")

    plt.tight_layout()
    plt.show()

visualize_results(model, val_dataset)
```
